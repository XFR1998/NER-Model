```python
import pandas as pd
from tqdm import tqdm
```

```python
KF=5
df_list = []
for i in range(1, KF+1):
    temp_df = pd.read_csv('./submit_{}.csv'.format(i))
    temp_df['tag'] = temp_df['tag'].apply(lambda x: eval(x))
    df_list.append(temp_df)
df1, df2, df3, df4, df5 = df_list
```

```python
len(df1),len(df2),len(df3),len(df4),len(df5)
```

```python

KF, (KF//2+1)
```

```python
all_tag_list = []


for i in tqdm(range(len(df1))):

    label1 = df1['tag'][i]
    label2 = df2['tag'][i]
    label3 = df3['tag'][i]
    label4 = df4['tag'][i]
    label5 = df5['tag'][i]

    true_entity = list(set(label1+label2+label3+label4+label5))

    all_tag_list.append(true_entity)
```

```python

all_tag_list[7].remove('wi')
```

```python
new_df = pd.DataFrame({'tag': all_tag_list})
new_df.to_csv('submit666.csv', index=False)
```
